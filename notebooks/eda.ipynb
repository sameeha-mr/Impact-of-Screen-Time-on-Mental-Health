```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv( "C:\\Users\\Sameeha\\New folder\\DataAnalysis\\data\\updated.csv")
```

```python
df['gender'].hist()
plt.show()
```

```python
df['phone_usage_hours'].hist()
plt.show()
```

```python
sns.scatterplot(x='daily_screen_time_hours', y='stress_level', data=df)
plt.show()
```

```python
plt.figure(figsize=(16, 14))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f", annot_kws={"size": 8})
plt.title('Correlation Matrix of Numeric Variables', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
corr_matrix = df.corr()

# Filter correlations above a threshold (e.g., 0.5)
strong_corr = corr_matrix[(corr_matrix > 0.5) | (corr_matrix < -0.5)]
print(strong_corr)
sns.heatmap(strong_corr.corr(), annot=True, cmap='coolwarm')
plt.show()
```
