```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv( "C:\\Users\\Sameeha\\New folder\\DataAnalysis\\data\\updated.csv")
```

```python
bins = [0, 2, 5, 10]  # Define intervals
labels = ['Low', 'Medium', 'High']
df['screen_time_category'] = pd.cut(df['daily_screen_time_hours'], bins=bins, labels=labels)
```

```python
sns.boxplot(x='screen_time_category', y='stress_level', data=df)
plt.show()
```
